```
#FIND MY RIDE

def find_my_car(make_type, miles, year, price):
  return cars[ (cars['Make'] == make_type) & (cars['Mileage'] <= miles) & (cars['Year'] == year) & (cars['Price'] <= price)].sort_values([ 'Year', 'Price'], ascending=[False,False]).head(25)

find_my_car('Ford',5000,2017,30000)              
                                                   
```
